# **Primera parte: Concatenar datos**

```python
import pandas as pd

# Cargar el primer archivo Excel
df1 = pd.read_excel("/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/input/agricola_2006_2018.xlsx")

# Cargar el segundo archivo Excel
df2 = pd.read_excel("/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/input/agricola_2019_2023.xlsx")
```

```python
print("Primer DataFrame:")
df1.head()
```

```python
print("\nSegundo DataFrame:")
df2.head()
```

```python
print("Columnas del primer DataFrame:")
print(df1.columns)

print("\nColumnas del segundo DataFrame:")
df2.columns
```

```python
# Filtrar columnas para df1
columns_to_keep_df1 = ['DEPARTAMENTO', 'MUNICIPIO', 'CULTIVO', 'AÑO', 'PERIODO', 'Área Sembrada\n(ha)', 'Área Cosechada\n(ha)', 'Producción\n(t)', 'Rendimiento\n(t/ha)', 'CICLO DE CULTIVO']
df1_filtered = df1[columns_to_keep_df1]

# Filtrar columnas para df2
columns_to_keep_df2 = ['Departamento', 'Municipio', 'Cultivo', 'Ciclo del cultivo', 'Año', 'Periodo', 'Área sembrada (ha)', 'Área cosechada (ha)', 'Producción (t)', 'Rendimiento (t/ha)']
df2_filtered = df2[columns_to_keep_df2]

# Renombrar columnas de df2 para que coincidan con df1 (opcional, pero recomendado)
df2_filtered = df2_filtered.rename(columns={
    'Departamento': 'DEPARTAMENTO',
    'Municipio': 'MUNICIPIO',
    'Cultivo': 'CULTIVO',
    'Ciclo del cultivo': 'CICLO DE CULTIVO',
    'Año': 'AÑO',
    'Periodo': 'PERIODO',
    'Área sembrada (ha)': 'Área Sembrada\n(ha)',
    'Área cosechada (ha)': 'Área Cosechada\n(ha)',
    'Producción (t)': 'Producción\n(t)',
    'Rendimiento (t/ha)': 'Rendimiento\n(t/ha)'
})

# Concatenar los DataFrames
df_concatenated = pd.concat([df1_filtered, df2_filtered], ignore_index=True)

# Exportar a Excel
output_path = "/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/output/datos_concatenados.xlsx"
df_concatenated.to_excel(output_path, index=False)
```

# **Segunda parte: Limpiar y exportar datos**

```python
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/output/datos_concatenados.xlsx")
df.head()
```

```python
# Obtiene los valores únicos de las columnas
departamentos_unicos = df["DEPARTAMENTO"].unique()
cultivos_unicos = df["CULTIVO"].unique()
ciclo_unicos = df["CICLO DE CULTIVO"].unique()

# **Crea DataFrames separados para "DEPARTAMENTO" y "CULTIVO"**
df_departamentos = pd.DataFrame({"DEPARTAMENTO": departamentos_unicos})
df_cultivos = pd.DataFrame({"CULTIVO": cultivos_unicos})
df_ciclos = pd.DataFrame({"CICLO DE CULTIVO": ciclo_unicos})

# Concatena los DataFrames
df_unicos = pd.concat([df_departamentos, df_cultivos, df_ciclos], axis=1)

# Guarda el nuevo DataFrame en un archivo Excel
df_unicos.to_excel("/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/output/valore_unicos.xlsx", index=False)
```

```python
# Convierte los valores de las  columnas "DEPARTAMENTO" y "CULTIVO" a mayúscula
df["DEPARTAMENTO"] = df["DEPARTAMENTO"].str.upper()
df["CULTIVO"] = df["CULTIVO"].str.upper()
# Define the mappings for departments and crops
departamento_mapping = {
    "ARCHIPIÉLAGO DE SAN ANDRÉS, PROVIDENCIA Y SANTA CATALINA": "SAN ANDRES Y PROVIDENCIA",
    "ATLANTICO": "ATLÁNTICO",
    "BOYACA": "BOYACÁ",
    "CAQUETA": "CAQUETÁ",
    "CHOCO": "CHOCÓ",
    "CORDOBA": "CÓRDOBA",
    "GUAINIA": "GUAINÍA",
    "QUINDIO": "QUINDÍO",
    "VAUPES": "VAUPÉS"
}

cultivo_mapping = {
    "AJI": "AJÍ",
    "AJONJOLI": "AJONJOLÍ",
    "ALGODON": "ALGODÓN",
    "ANON": "ANÓN",
    "ARANDANO": "ARÁNDANO",
    "ARAZA": "ARAZÁ",
    "ASAI": "ASAÍ",
    "BOROJO": "BOROJÓ",
    "BROCOLI": "BRÓCOLI",
    "CAFE": "CAFÉ",
    "CALABACÍN, CALABAZA": "CALABACIN",
    "CAÑA AZUCARERA": "CAÑA",
    "CAÑA FLECHA": "CAÑA",
    "CAÑA MIEL": "CAÑA",
    "CAÑA PANELERA": "CAÑA",
    "CIMARRON": "CIMARRÓN",
    "COPOAZU": "COPOAZÚ",
    "CURCUMA": "CÚRCUMA",
    "CÚRCUMA O AZAFRÁN": "CÚRCUMA",
    "DATIL": "DÁTIL",
    "DURAZNO O ALBARICOQUE": "DURAZNO",
    "ESPARRAGO": "ESPÁRRAGO",
    "GUANABANA": "GUANÁBANA",
    "LIMON": "LIMÓN",
    "MAIZ": "MAÍZ",
    "MAIZ FORRAJERO": "MAÍZ",
    "MALANGA, ACHÍN, YOTA, PAPA CHINA, BORE": "MALANGA",
    "MANI": "MANÍ",
    "MARACUYA": "MARACUYÁ",
    "MARAÑON": "MARAÑÓN",
    "MELON": "MELÓN",
    "NISPERO": "NÍSPERO",
    "OREGANO": "ORÉGANO",
    "PAPAYUELA -BABACO": "PAPAYUELA",
    "PIMENTON": "PIMENTÓN",
    "PLANTAS AROMATICAS": "PLANTAS AROMÁTICAS",
    "PLANTAS AROMATICAS Y MEDICINALES": "PLANTAS AROMÁTICAS",
    "PLATANO": "PLÁTANO",
    "RABANO": "RÁBANO",
    "RAMBUTAN": "RAMBUTÁN",
    "SABILA": "SÁBILA",
    "TANGELO": "TÉ",
    "TOMATE DE ARBOL": "TOMATE DE ÁRBOL"
}


ciclo_mapping = {
    "ANUAL": "PERMANENTE",
    "Permanente": "PERMANENTE",
    "Transitorio": "TRANSITORIO"
}

# Rename values in 'DEPARTAMENTO' column
df["DEPARTAMENTO"] = df["DEPARTAMENTO"].replace(departamento_mapping)

# Rename values in 'CULTIVO' column
df["CULTIVO"] = df["CULTIVO"].replace(cultivo_mapping)

# Rename values in 'CULTIVO' column
df["CICLO DE CULTIVO"] = df["CICLO DE CULTIVO"].replace(ciclo_mapping)


df = df.rename(columns={
    'Area Sembrada (ha)': 'AREA SEMBRADA',
    'Area Cosechada (ha)': 'AREA COSECHADA',
    'Produccion (t)': 'PRODUCCION',
    'Rendimiento (t/ha)': 'RENDIMIENTO'
})

# Export the modified DataFrame to Excel
df.to_excel("/content/drive/MyDrive/GITHUB/COLAB/cultivos_colombia/output/datos_2006_2023.xlsx", index=False)
```
